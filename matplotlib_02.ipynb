# BEAUTIFICATION OF DATA

```python
import matplotlib.pyplot as plt
```

```python
import numpy as np
```

```python
x = np.linspace(0,10, 200)
```

```python
x
```

```python
y = np.sin(x)
```

```python
y
```

```python
plt.figure(figsize = (6,2))
plt.plot(x,y, '--b')
plt.xlabel("this is my x data ")
plt.ylabel("this is my y data ")
plt.title("this is my title")
plt.grid()
plt.show()
```

```python
x = np.random.rand(50)
y = np.random.rand(50)

colours = np.random.rand(50)
sizes = 1000*np.random.rand(50)
plt.scatter(x,y, c = "#A64D13", s = sizes, alpha=.5, marker= 'v')
plt.xlabel("x axis")
plt.ylabel("y axis")
plt.title("random data")
```

```python
x = [4,5,6,7,8,9]
y = [6,7,8,9,10,11]
plt.plot(x,y)
plt.show()
```

```python
x = [4,5,6,7,8,9]
y = [60,30,80,19,40,1]
plt.plot(x,y)
plt.show()
```

```python
x = [4,5,6,7,8,9]
y = [60,30,80,19,40,1]
plt.scatter(x,y)
plt.show()
```

```python
x = [4, 5, 6, 7, 8, 9]
y = [60, 30, 80, 10, 40, 1]
colors = ['red', 'blue', '#4E301D', '#DFC010', '#EC801C', '#0F6AEF']  

plt.scatter(x, y, c=colors, s=100)
plt.show()
```

```python
x = np.random.rand(50)
y = np.random.rand(50)
z = np.random.rand(50)
```

```python
fig = plt.figure()
ax = fig.add_subplot(projection = '3d')
ax.scatter(x,y,z)
plt.show()
```
